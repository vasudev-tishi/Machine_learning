```python
# droping
df=df.drop('Extracurricular Activities',axis=1)
```

```python
# Correlation matrix using scatter plot
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.scatterplot(correlation_matrix)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plotting the correlation matrix with heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
for column in df.select_dtypes(include=['int64', 'float64']):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
# Check the distribution of a numerical feature after cleaning
for column in df.select_dtypes(include=['int64', 'float64']):
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column} (After Cleaning)')
    plt.show()
```

```python
# Analyze numerical features after outlier removal
for column in df.select_dtypes(include=['int64', 'float64']):
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column} (After Outlier Removal)')
    plt.show()

    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=20)  # Adjust 'bins' as needed
    plt.title(f'Histogram of {column} (After Outlier Removal)')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

    plt.figure(figsize=(8, 6))
    df[column].value_counts().plot(kind='bar')
    plt.title(f'Bar plot of {column} (After Outlier Removal)')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.boxplot(x=column, data=df)
    plt.title(f'Box Plot of {column} (After Outlier Removal)')
    plt.show()
```
